```python
%matplotlib inline

sns.set(font_scale=1.2)
```

```python
deltas1 = [np.random.uniform(0.0, 1.0) for _ in range(50)]
deltas2 = [np.random.uniform(0.0, 1.0) for _ in range(50)]
deltas3 = [np.random.uniform(0.0, 1.0) for _ in range(50)]
```

```python
timesteps = [i for i in range(1, 51)]
```

```python
single_dev_inc = [0.0]

for ts in timesteps:
    change = single_dev_inc[ts-1] + deltas1[ts-1]
    single_dev_inc.append(change)
```

```python
team_dev_inc = [0.0]
```

```python
for ts in timesteps:
    change1 = team_dev_inc[ts-1] + (team_dev_inc[ts-1] + deltas1[ts-1])
    change2 = team_dev_inc[ts-1] + (team_dev_inc[ts-1] + deltas2[ts-1])
    change3 = team_dev_inc[ts-1] + (team_dev_inc[ts-1] + deltas3[ts-1])
    team_dev_inc.append(change1 + change2 + change3)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

plt.tight_layout(pad=4.0, w_pad=5., h_pad=0.25)
plt.suptitle("Disk space requirements with ad hoc versioning", fontsize="18")
ax1.plot(single_dev_inc)
ax1.set_title("Single Developer")
ax1.set_xlabel("Num. Changes")
ax1.set_ylabel("Disk space");

ax2.set_title("Team with 3 members")
ax2.plot(team_dev_inc[0:5])
ax2.set_xlabel("Num. Changes per Modeler");

plt.savefig('{}/disk_usage.png'.format(FIG_DIR), format='png', dpi=300)
```
